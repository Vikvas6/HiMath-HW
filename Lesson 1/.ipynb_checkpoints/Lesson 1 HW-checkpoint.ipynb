Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
def vect_abs(x, y, z):
    return np.sqrt(x**2 + y**2 + z**2)
```

### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
def plot_circle(r=1):
    x = np.linspace(-r, r, 1000)
    y_up = np.sqrt(abs(r**2 - x**2))
    plt.plot(x, y_up, color='b')
    plt.plot(x, -y_up, color='b')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

```python
plot_circle()
```

```python
def plot_ellipse(a=2, b=1):
    x = np.linspace(-a, a, 1000)
    y_up = np.sqrt(abs(b**2*(1 - x**2/a**2)))
    plt.plot(x, y_up, color='b')
    plt.plot(x, -y_up, color='b')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

```python
plot_ellipse()
```

```python
def plot_hyperbola(a=1, b=2):
    xl = []
    xr = []
    y1 = []
    y2 = []
    for i in range(100):
        x_ = i/100 + a
        xl.append(-x_)
        xr.append(x_)
        y1.append(np.sqrt(abs(-b**2*(1 - x_**2/a**2))))
        y2.append(-np.sqrt(abs(-b**2*(1 - x_**2/a**2))))
    plt.plot(xl, y1, color='b')
    plt.plot(xl, y2, color='b')
    plt.plot(xr, y1, color='b')
    plt.plot(xr, y2, color='b')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

```python
plot_hyperbola()
```

### 5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.
Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
from mpl_toolkits.mplot3d import Axes3D
from pylab import *
```

```python
def parallel(a=1, b=1, c=30):
    fig = figure()
    ax = Axes3D(fig)
    X = np.arange(-5, 5, 0.5)
    Y = np.arange(-5, 5, 0.5)
    X, Y = np.meshgrid(X, Y)
    Z = a*X + b*Y
    Z2 = a*X + b*Y + c
    ax.plot_wireframe(X, Y, Z)
    ax.plot_wireframe(X, Y, Z2, color='g')
    show()
```

```python
parallel()
```

```python
def second_order(a=1, b=1, c=30):
    fig = figure()
    ax = Axes3D(fig)
    X = np.arange(-5, 5, 0.5)
    Y = np.arange(-5, 5, 0.5)
    X, Y = np.meshgrid(X, Y)
    Z = a*X + b*Y**2
    Z2 = a*X**2 + b*Y**2 + c
    ax.plot_wireframe(X, Y, Z)
    ax.plot_wireframe(X, Y, Z2, color='g')
    show()
```

```python
second_order()
```

### 3. Задание (в программе)

Напишите код, который будет рисовать график окружности в полярных координатах.

```python
def polar_to_cartesian(r, a):
    return r * np.cos(a), r * np.sin(a)
```

```python
def plot_circle_polar(r):
    a = np.linspace(0, 360, 1000)
    X, Y = polar_to_cartesian(r, a)
    plt.plot(X, Y)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

```python
plot_circle_polar(1)
```

### 4. Задание (в программе)
Решите систему уравнений:

y = x^2 – 1

exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, x**2-1)
plt.plot(x, 1-(1-np.exp(x))/x)
plt.grid(True)
plt.show()
```

```python
def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)
```

```python
x1, y1 = fsolve(equations, (-1.5, 2))
print(x1, y1)
x1, y1 = fsolve(equations, (2.5, 6))
print(x1, y1)
x1, y1 = fsolve(equations, (4, 17))
print(x1, y1)
```

Решите систему уравнений и неравенств:

y = x^2 – 1

exp(x) + x∙(1 – y) > 1

```python
# y = x^2 – 1
def equation(x):
    return x**2 - 1
```

```python
# y < (exp(x) + x - 1) / x
def inequation(x):
    return (x + np.exp(x) - 1)/x
```

```python
# x^2 – 1 < (exp(x) + x - 1) / x
# 0 < (exp(x) + x - 1) / x - x^2 – 1
def subtraction(x):
    return inequation(x) - equation(x)
```

```python
x = np.linspace(-3, 5, 100)
plt.plot(x, subtraction(x))
plt.grid(True)
plt.show()
```

По графику определяем, что итоговое неравенство положительно от первого корня до второго и от третьего корня до бесконечности (всё не включительно)

```python
x1_s = fsolve(subtraction, -1.5)[0]
x1_e = fsolve(subtraction, 2.5)[0]
x2_s = fsolve(subtraction, 4)[0]

print(f"({x1_s}, {x1_e}), ({x2_s}, +Inf)")
```
